```python
cp2
```

```python
cp_final = cp2.drop(['Car Model', 'Mercedez Benz C class'], axis="columns")
cp_final
```

```python
X_cp = cp_final.drop(['Sell Price($)'], axis="columns")
X_cp
```

```python
Y_cp = cp_final['Sell Price($)']
Y_cp
```

```python
model_cp = LinearRegression()
model_cp.fit(X_cp.values, Y_cp.values)
```

```python
model_cp.predict([[45000, 4, 0, 0]])
```

```python
model_cp.predict([[86000, 7, 0, 1]])
```
