```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
model = DecisionTreeClassifier()
```

```python
salaries = pd.read_csv('data/salaries.csv')
salaries.head()
```

```python
np.unique(salaries.company)
```

```python
np.unique(salaries.job)
```

```python
np.unique(salaries.degree)
```

```python
input = salaries.drop('salary_more_then_100k', axis='columns')
target = salaries.salary_more_then_100k
le_company = LabelEncoder()
input['company'] = le_company.fit_transform(input['company'])
le_job = LabelEncoder()
input['job'] = le_job.fit_transform(input['job'])
le_degree = LabelEncoder()
input['degree'] = le_degree.fit_transform(input['degree'])
input.head()
```

```python
input.shape
```

```python
plt.scatter(salaries.degree, salaries.salary_more_then_100k, color='blue')
plt.scatter(salaries.job, salaries.salary_more_then_100k, color='red')
plt.scatter(salaries.company, salaries.salary_more_then_100k, color='green')
plt.show()
```

```python
model.fit(input.values, target)
```

```python
model.predict([[2, 1, 1]])
```

```python
model.predict([[0, 0, 0]])
```

#### Excercise

```python
titanic_df = pd.read_csv('data/titanic.csv')
titanic_df.head()
```

```python
titanic_df.shape
```

```python
input = titanic_df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
input.head()
```

```python
input.Sex = input.Sex.map({'male': 1, 'female': 2})
input.head()
```

```python
input.isna().any()
```

```python
input.Age = input.Age.fillna(input.Age.mean())
input.head()
```

```python
input.Age[0:10]
```

```python
X_train, X_test, y_train, y_test = train_test_split(input.drop('Survived', axis='columns'), input.Survived, test_size=0.2) 
```

```python
X_train.shape
```

```python
model_titanic = DecisionTreeClassifier()
model_titanic.fit(X_train, y_train)
```

```python
model_titanic.score(X_test, y_test)
```
