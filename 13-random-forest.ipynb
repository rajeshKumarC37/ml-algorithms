```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
digits_data = load_digits()
dir(digits_data)
```

```python
digits_data.data[0]
```

```python
digits_data.feature_names[0]
```

```python
digits_data.target
```

```python
digits_data.target_names
```

```python
digits_df = pd.DataFrame(digits_data.data)
digits_df.head()
```

```python
digits_df['target'] = digits_data.target
```

```python
digits_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits_df.drop('target', axis='columns'), digits_df.target, test_size=0.2)
```

```python
X_train.shape
```

```python
model = RandomForestClassifier(n_estimators=50)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.datasets import load_iris
iris_data = load_iris()
dir(iris_data)
```

```python
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris_df.head()
```

```python
iris_df['target'] = iris_data.target
iris_df['flower'] = iris_df.target.apply(lambda x: iris_data.target_names[x])
iris_df.head()
```

```python
df1 = iris_df[:50]
df2 = iris_df[50:100]
df3 = iris_df[100:150]
df2.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_df.drop(['target', 'flower'], axis='columns'), iris_df.target, test_size=0.2)
```

```python
X_train.shape
```

```python
model_rf = RandomForestClassifier(n_estimators=50)
model_rf.fit(X_train, y_train)
```

```python
model_rf.score(X_test, y_test)
```

```python
model_rf2 = RandomForestClassifier(n_estimators=10)
model_rf2.fit(X_train, y_train)
```

```python
model_rf2.score(X_test, y_test)
```
