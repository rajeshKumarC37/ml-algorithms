```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from math import floor
```

```python
df = pd.read_csv("data/homeprices-multiple-var.csv")
df
```

```python
df.bedrooms = df.bedrooms.fillna(floor(df.bedrooms.median()))
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop("price", axis="columns").values, df.price)
```

```python
reg.predict([[3400, 4, 10]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
def ifNanReturnZero(item):
    if(isinstance(item, float)):
        if(math.isnan(item)):
            return 0
    return item

def convertStringToNumber(item):
    if(isinstance(item, str)):
        return w2n.word_to_num(item)
    else:
        return ifNanReturnZero(item)

# convertStringToNumber(float("NaN"))
df2.experience = df2.experience.fillna("zero")
df2.experience = df2.experience.apply(w2n.word_to_num)
df2.test_score = df2.test_score.fillna(math.floor(df2.test_score.mean()))
# df2.experience = [convertStringToNumber(item) for item in df2.experience]
# df2.test_score = [ifNanReturnZero(item) for item in df2.test_score]
# df2.interview_score = [ifNanReturnZero(item) for item in df2.interview_score]
df2
```
